```python
#Q.1
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

driver=webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ")
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python
#name
loc_tags=driver.find_elements(By.XPATH,'//tbody/tr/td[1]/a')
for i in loc_tags:
    location=i.text
    name.append(location) 
    
#artist   
dt_tags=driver.find_elements(By.XPATH,'//tbody/tr/td[2]/a')
for i in dt_tags:
    if i.text is None :
        artist.append("-")
    else:
        artist.append(i.text)
        
#upload date
tm_tags=driver.find_elements(By.XPATH,'//tbody/tr/td[4]')
for i in tm_tags:
    if i.text is None :
        upload_date.append("-") 
    else:
        upload_date.append(i.text)
    
#views
dt_tags=driver.find_elements(By.XPATH,'//tbody/tr/td[3]')
for i in dt_tags:
    if i.text is None :
        views.append("-")
    else:
        views.append(i.text)
```

```python
print(len(name),len(artist),len(upload_date),len(views))
```

```python
df=pd.DataFrame([])
df['Name']=name[:38]
df['Artist']=artist[:38]
df['Upload Date']=upload_date[:38]
df['Views']=views[:38]
df
```

```python
#Q.8
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/ ")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a")
search.click() #click on dataset
```

```python
#fetching url to open each dataset
url_list=[]
url=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a')
for i in url:
    url_list.append(i.get_attribute('href'))
```

```python
len(url_list)
```

```python
url_list
```

```python
dataset_name=[]
data_type=[]
task=[]
att_type=[]
no_instance=[]
no_att=[]
year=[]
```

```python
for i in url_list:# iterating each and every url to fetch full datasets details
    driver.get(i) 
    time.sleep(5)
    try: #dataset name
        brnd=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1')
        dataset_name.append(brnd.text)
    except NoSuchElementException :
        dataset_name.append('-')
        
    try: #data type
        cl=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        data_type.append(cl.text)
    except NoSuchElementException :
        data_type.append('-')
        
    try: #task
        re=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        task.append(re.text)
    except NoSuchElementException :
        task.append('-')
        
    try: #attribute type
        brnd=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        att_type.append(brnd.text)
    except NoSuchElementException :
        att_type.append('-')
        
    try: # no of instances
        sp=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        no_instance.append(sp.text)
    except NoSuchElementException :
        no_instance.append('-')
        
    try: #no of attribute
        cl=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        no_att.append(cl.text)
    except NoSuchElementException :
        no_att.append('-')
        
    try: #year
        re=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        year.append(re.text)
    except NoSuchElementException :
        year.append('-')
```

```python
print(len(dataset_name),len(data_type),len(task),len(att_type),len(no_instance),len(no_att),len(year))
```

```python
df=pd.DataFrame({'Dataset Name':dataset_name,'DataType':data_type,'Task':task,'Attribute Type':att_type,'No_Of_Instances':no_instance,'No_Of_Attribute':no_att,'Year':year})
df
```
