```python
rating=[]
review_sum=[]
full_rew=[]
```

```python
#Scraping rating from given page
rat_tags=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
for i in rat_tags:
    rats=i.text
    rating.append(rats)
    
#Scraping Review summary from given page
sum_tags=driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
for i in sum_tags:
    summary=i.text
    review_sum.append(summary)    
    
#Scraping full summary from given page
rev_tags=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')
for i in rev_tags:
    rev_sum=i.text
    full_rew.append(rev_sum)
```

```python
rats=[]
revs=[]
revf=[]
```

```python
print(len(rating),len(review_sum),len(full_rew))
```

```python
start=0
end=10
for page in range(start,end):
    ratings=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in ratings:
        rats.append(i.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')#to scrape data from next pages as well
    next_button.click()
    time.sleep(3)
```
